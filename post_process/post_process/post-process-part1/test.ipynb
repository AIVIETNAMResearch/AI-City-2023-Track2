```python
import torch
import os
import json
sim_mat = torch.load('./sim_mat/rescore_sim_mat_soup_wmeg.pth', map_location=torch.device('cpu'))

submit_dict = dict()

# load the info of the sim_mat 
test_tracks = json.load(open("./data/test_tracks.json"))
test_track_ids = list(test_tracks.keys())

test_querries = json.load(open("./data/test_queries.json"))
test_query_ids = list(test_querries.keys())

rank_res = (-sim_mat).argsort()  # query_ids x track_ids
rank_res = rank_res.astype(int)
for i,query_id in enumerate(test_query_ids):
    submit_dict[query_id] = list(np.array(test_track_ids)[rank_res[i,:]])

with open(os.path.join("./results","submit.json"),'w') as f:
    json.dump(submit_dict,f)
```
